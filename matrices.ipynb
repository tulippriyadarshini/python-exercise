```python
import numpy as np
mydata = np.arange(0,20)
print(mydata)
```

```python
np.reshape(mydata,(5,4))
```

```python
matrix1 = np.reshape(mydata,(5,4), order='C')
print(matrix1)
```

```python
print(matrix1[2,2])
```

```python
matrix2 = np.reshape(mydata,(5,4), order='F')
print(matrix2)
```

```python
print(matrix2[0,2])
```

```python
#OOP Concept:
mydata.reshape((5,4))
```
