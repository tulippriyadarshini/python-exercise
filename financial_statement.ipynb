```python
revenue = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
```

```python
expenses = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]
```

```python
expenses
```

```python
total_revenue = sum(revenue)
total_expenses = sum(expenses)

mean_revenue = total_revenue/len(revenue)
mean_expenses = total_expenses/len(expenses)

max_profit = 0
min_profit = 0

for i in range(len(revenue)):
   diff = revenue[i] - expenses[i]
   diff_after_tax = diff-diff*0.3
   profit_after_tax = diff_after_tax/revenue[i]
   if diff_after_tax> max_profit:
      max_profit = diff_after_tax  
   if  diff_after_tax < min_profit:
      min_profit = diff_after_tax
   print("Profit/Loss", round(diff))
   print("Profit/Loss after tax",round(diff_after_tax))
   print("Profit/Loss Margin", profit_after_tax)
   print(round(mean_revenue))
   if profit_after_tax > mean_revenue:
      print("Good Months")
   else:
      print("Bad Months")
print("Best Month Profit", max_profit)
print("Worst Month Profit", min_profit)
```

```python
import numpy as np

expenses_arr = np.array(expenses)
revenue_arr = np.array(revenue)

profit_arr = revenue_arr-expenses_arr
profit_after_tax_arr = profit_arr - profit_arr*0.3
profit_margin_arr = profit_after_tax_arr/revenue_arr
profit_after_tax_arr_mean = profit_after_tax_arr.mean()
print(profit_arr)
print(profit_after_tax_arr)
print(profit_margin_arr)
print(profit_after_tax_arr_mean)

for profit_after_tax_1 in profit_after_tax_arr:
    if profit_after_tax_1 > profit_after_tax_arr_mean:
        print("Good Months")
    else:
        print("Bad Months")
print(profit_after_tax_arr.max())
print(profit_after_tax_arr.min())
```
