```python
str = "Python Programming"
count = 0
for char in str:
    if char.lower() in "aeiou":
        count = count+1
print("No. of vowel in", str, "is:", count)
```

```python
str = "Python Programming"
count = 0
vowels_in_str = [char for char in str if char.lower() in "aeiou"]
print("No. of vowel in", str, "is:", len(vowels_in_str))
```

```python
#Palindrome
palin_str = input("Enter String:")
rev_str = palin_str[::-1]
if palin_str == rev_str:
    print("Palindrome string")
else:
    print("Not Palindrome")
```

```python
str1 = "This is Python. Python is fun!"
word_breakers = [' ', ',', '.', '!']

words = []
single_word = ''

for char in str1:

    if char in word_breakers and len(single_word) != 0:
        words.append(single_word)
        single_word = ''
    else:
        single_word = (single_word + char).strip()

print(words)

counter=0

for i in words:
    if i.lower() == 'is':
        counter=counter+1
print(counter)

```
