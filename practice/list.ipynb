```python
duplicate_list = [1, 2, 5, 1, 2, 0, -1, -2, -1,5]
non_duplicate =[]
for item in duplicate_list:
    if not item in non_duplicate:
        non_duplicate.append(item)
print(non_duplicate)
```

```python
seen = set()
new_non_dup= [item for item in duplicate_list if not (item in seen or seen.add(item))]

print(new_non_dup)
```

```python
unique_list = list(set(duplicate_list))
print(duplicate_list)
print(unique_list)
```

```python
unique_list1 = list(dict.fromkeys(duplicate_list))
print(unique_list1)
```
