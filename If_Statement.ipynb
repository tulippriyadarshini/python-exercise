```python
#------- -2 ------- -1 ----- 0 ------- 1 ------- 2

x =  randn()
answer = None
if x>1:
    answer = "Greater than 1"
elif x>-1 and x<1:
    answer = "Between 1 and -1"
else:
    answer = "Less than 1"
    
print(x)
print(answer)
```
